```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
np.random.seed(0)
X1= np.random.uniform(0, 1, 5000) - 0.5
X2 = np.random.uniform(0, 1, 5000) - 0.5
y1 = 1*(X1**2 - X2**2 > 0)
y2 = 1*(np.abs(X1) + np.abs(X2) > 0.5)
y3 = 1*(X1**2 + X2**2 > 0.2)
y4 = 1*(2*X1**2 + 0.5*X2**2 > 0.1)
```

```python
data = pd.DataFrame({'X1':X1, 'X2':X2, 'y1':y1,'y2':y2,'y3':y3,'y4':y4})
```

```python
fig,ax = plt.subplots(2,2,figsize = (8,8))
sns.scatterplot(x='X1', y='X2', hue='y3', data=data,ax=ax[0][0])
sns.scatterplot(x='X1', y='X2', hue='y1', data=data,ax=ax[0][1])
sns.scatterplot(x='X1', y='X2', hue='y2', data=data,ax=ax[1][0])
sns.scatterplot(x='X1', y='X2', hue='y4', data=data,ax=ax[1][1])
for t in ax[0]:
    t.set_xticks([])
    t.set_yticks([])
    plt.setp(t.get_legend().get_texts(), fontsize='1') # for legend text
    plt.setp(t.get_legend().get_title(), fontsize='1') # for legend title
    
for t in ax[1]:
    t.set_xticks([])
    t.set_yticks([])    
    plt.setp(t.get_legend().get_texts(), fontsize='1') # for legend text
    plt.setp(t.get_legend().get_title(), fontsize='1') # for legend title
```

```python
np.random.seed(0)
X1= np.random.uniform(-1,1, 10000) 
X2 = np.random.uniform(-1, 1, 10000)
y1 = 1*( (X1**2 + X2**2 > 0.2) & (X1**2 + X2**2 < 0.8) )
data = pd.DataFrame({'X1':X1, 'X2':X2, 'y1':y1})
plt.figure(figsize = (6,6))
sns.scatterplot(x='X1', y='X2', hue='y1', data=data)
plt.xticks([])
plt.yticks([])
```
